```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
mnist = load_digits()
data = mnist.data / 255.
```

```python
def plot_components(x, y, target, title=None):
    f, ax = plt.subplots(1,1)
    mpl.rcParams['figure.figsize'] = 11.7,8.27
    ax = sns.scatterplot(x, y, hue=target, palette=sns.color_palette(), alpha=0.7)
    ax.set_title(title)
    ax.set_xlabel("Component 1")
    ax.set_ylabel("Component 2");
```

```python
pca = PCA(n_components=2)
pca_fit = pca.fit(data)
pca_transformed = pca_fit.transform(data)

print("Variance explained by first two PCAs: {:.1f}%".format(sum(pca_fit.explained_variance_ratio_)*100))
x, y = pca_transformed[:,0], pca_transformed[:,1]
target = mnist.target 

plot_components(x, y, target, title="First two principle-components of MNIST")
```

```python
tsne = TSNE(n_components=2)
tsne_transformed = tsne.fit_transform(data)
x, y = tsne_transformed[:,0], tsne_transformed[:,1]

plot_components(x, y, mnist.target, title="TSNE components")
```
